```python
columns_to_convert = df.select_dtypes(include=['object', 'float64']).columns


for col in columns_to_convert:
    try:
        
        df[col] = df[col].fillna(0)  
        df[col] = df[col].astype(int)
    except:
        print(f"Column '{col}' cannot be converted to integer.")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Cabin'], errors='ignore')
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
df.isnull().sum()
```

Missing Values Analysis

```python
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

Univariate Analysis


 Numerical Distributions

```python
numerical_cols = ['Age', 'Fare']
df[numerical_cols].hist(bins=20, figsize=(10, 4), layout=(1, 2), color='skyblue')
plt.suptitle("Numerical Feature Distributions")
plt.show()
```

Categorical Features

```python
categorical_cols = ['Sex', 'Pclass', 'Embarked']
for col in categorical_cols:
    sns.countplot(data=df, x=col)
    plt.title(f"Count Plot for {col}")
    plt.show()
```

Sex: More males than females.

Pclass: Majority in 3rd class.

```python
#Outlier Detection


sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()
```

```python
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
plt.show()
```

Fare: High-end outliers (up to $500+).

Age: Reasonable spread, few old-age outliers (70+).



```python
# Bivariate Analysis


# Bivariate Analysis



sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

```

```python
 # Survival by Category

sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Sex")
plt.show()

sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

Multivariate Analysis

```python
# Age vs Fare by Survival

sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare (colored by Survival)")
plt.show()
```

Violin Plot: Age Distribution by Class & Survival

```python
sns.violinplot(x="Pclass", y="Age", hue="Survived", data=df, split=True)
plt.title("Age Distribution by Pclass and Survival")
plt.show()
```

Younger passengers in 3rd class had low survival.

Older passengers in 1st class had better odds.
